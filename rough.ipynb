```python
import os
import shutil

def copy_files_to_folder(source_dir, dest_dir):
    """
    Copies all files from a directory and its subdirectories to another directory.

    Args:
        source_dir (str): The path to the directory containing the files to copy.
        dest_dir (str): The path to the directory where the files will be copied.
    """
    # Ensure that the destination directory exists.
    if not os.path.exists(dest_dir):
        try:
            os.makedirs(dest_dir)
        except OSError as e:
            print(f"Error creating destination directory {dest_dir}: {e}")
            return

    # Walk through the source directory.
    for root, _, files in os.walk(source_dir):
        for filename in files:
            # Construct the full path to the file.
            src_file_path = os.path.join(root, filename)
            dest_file_path = os.path.join(dest_dir, filename) # keep original filename
            # Copy the file to the destination directory.
            try:
                shutil.copy2(src_file_path, dest_file_path)  # copy metadata too
                print(f"Copied: {src_file_path} to {dest_file_path}")
            except (OSError, shutil.Error) as e:
                print(f"Error copying file {src_file_path} to {dest_file_path}: {e}")
                # Consider if you want to continue copying other files
                # even if one fails.  If so, just 'continue' here.
                # If you want the whole process to stop on any error,
                # then 'return' here.
                continue # keep copying other files

    print("File copy process completed.")

if __name__ == "__main__":
    # Get the source and destination directories from the user.
    source_directory = '/Users/shresth/Documents/Anonymous-SAIL'
    destination_directory = '/Users/shresth/Documents/Anonymous-SAIL-full'

    # Call the function to copy the files.
    copy_files_to_folder(source_directory, destination_directory)
```
